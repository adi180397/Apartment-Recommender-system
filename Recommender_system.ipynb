```python
import re

def extract_sector(df, column_name):
    # Define a regex pattern to match 'Sector <number>'
    pattern = r"Sector \d+"

    # Use the apply function to extract the sector using regex
    df['sector'] = df[column_name].apply(lambda x: re.search(pattern, x).group(0) if re.search(pattern, x) else None)

    return df
```

```python
import ast

def clean_string(facility_string):
    # Safely evaluate the string representation of a list
    try:
        # Convert string representation of list to actual list
        facility_list = ast.literal_eval(facility_string)
        # Join the list into a single string
        return ' '.join(facility_list)
    except (ValueError, SyntaxError):
        # Handle any errors in case of malformed strings
        return facility_string

# Apply cleaning function
merge_df['TopFacilities_str'] = merge_df['TopFacilities'].apply(clean_string)
```

**Embeddings**

```python
from transformers import BertTokenizer, BertModel
import torch
```

```python
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
model = BertModel.from_pretrained('bert-base-uncased')
```

```python
def get_bert_embedding(text):
    inputs = tokenizer(text, return_tensors='pt', truncation=True, padding=True)
    with torch.no_grad():
        outputs = model(**inputs)
    return outputs.last_hidden_state.mean(dim=1).squeeze().numpy()
```

```python
merge_df['BERT_Embeddings'] = merge_df['TopFacilities_str'].apply(lambda x: get_bert_embedding(x))
```

```python
merge_df.head()
```

```python
embeddings_matrix = np.stack(merge_df['BERT_Embeddings'].values)
```
